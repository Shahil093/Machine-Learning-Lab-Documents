```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings 
warnings.filterwarnings("ignore") 
```

```python
data = pd.read_csv('Expri_Dataset (1).csv') 
```

```python
data.info
```

```python
data.describe()
```

```python
from sklearn.preprocessing import Normalizer 
```

```python
x = data.iloc[: , :-1] 
```

```python
y = data.iloc[:, -1] 
```

```python
x.head() 
```

```python
columns = x.columns
```

```python
norm = Normalizer() 
```

```python
x_norm = norm.fit_transform(x) 
```

```python
x_norm = pd.DataFrame(x_norm, columns = columns)
```

```python
x_norm.head() 
```

```python
from sklearn.cluster import KMeans 
```

```python
data = list(zip(x)) 
```

```python
inertias = [] 
```

```python
for i in range(1,10):     
  kmeans = KMeans(n_clusters=i)     
  kmeans.fit(x_norm) 
  inertias.append(kmeans.inertia_) 
```

```python
plt.plot(range(1,10), inertias, marker='o') 
plt.title('Elbow method')
plt.xlabel('Number of clusters') 
plt.ylabel('Inertia') 
plt.show() 
```

```python
x.corr().style.background_gradient(cmap = 'coolwarm') 
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components = 2) 
pca.fit(x_norm) 
x_pca = pca.transform(x_norm)  
x_pca = pd.DataFrame(x_pca, columns = ['Feature_1', 'Feature_2']) 
x_pca.head() 
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
y = label_encoder.fit_transform(y) 
y = pd.DataFrame(y, columns = ['Target']) 
```

```python
x_pca = pd.concat([x_pca, y], axis = 1) 
```

```python
sns.scatterplot(data = x_pca, x='Feature_1', y='Feature_2', hue = 'Target')
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_pca, x_pca.iloc[:, -1], test_size=0.3, random_state=0)
```

```python
from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans.fit(X_train) 
```

```python
sns.scatterplot(data = X_train, x = 'Feature_1', y = 'Feature_2', hue = kmeans.labels_)
```

```python
from sklearn.metrics import silhouette_score 
silhouette_score(X_train, kmeans.labels_, metric='euclidean')
```

```python
K = range(2, 8) 
fits = [] 
score = [] 
for k in K: 
    model = KMeans(n_clusters = k, random_state = 0).fit(X_train)     
    fits.append(model) 
    score.append(silhouette_score(X_train, model.labels_, metric='euclidean')) 
sns.lineplot(x = K, y = score) 
```

```python
from sklearn import metrics 
labels = kmeans.labels_ 
Sil = metrics.silhouette_score(X_train, labels) 
CH = metrics.calinski_harabasz_score(X_train, labels) 
DB = metrics.davies_bouldin_score(X_train, labels) 
```
